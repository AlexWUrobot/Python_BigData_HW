WordCount : 請依據上列輸出的完整歌詞進⾏斷字，計算每個單字各別出
現次數，並排序輸出其結果 (參考 [提⽰] )。注意：輸出結果必須對⿑！

```
lyric_split = []
new_lyric_split = []
for sentence in output:
  lyric_split += sentence.split() # 以空格为分隔符  
lyric_split.sort()

for word in lyric_split[:]:
  newword = word.rstrip(',').rstrip('?')  # 移除分號
  new_lyric_split.append(newword) 
lyric_split = new_lyric_split

last_word = lyric_split[0]
count = 0
count_list = []
for word in lyric_split:
  if word == last_word:
    count += 1
  else:
    count_list.append([last_word,count])
    count = 1
  last_word = word
count_list.append([last_word,count])  # 加最後一個字

for [word,count] in count_list:
  lenword = len(word)
  print(word,end='')
  while( lenword < 20):
    print(' ',end='')
    lenword += 1
  print(count)

  #print("{0:<0} \t {1:>10}".format(word,count))  
```
